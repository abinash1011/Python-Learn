```python
import random
def max_pairwise_product(numbers):
    n = len(numbers)
    max_product = 0
    for first in range(n):
        for second in range(first + 1, n):
            max_product = max(max_product,
                numbers[first] * numbers[second])

    return max_product

def max_pairwise_product_quick(numbers):
    numbers = sorted(list(set(numbers)), reverse=True)
    return numbers[0] * numbers[1]


if __name__ == '__main__':
    counter = 0
    while counter < 10:
      input_n = random.randint(2, 1000)
      input_numbers = random.sample(range(0, 100000), input_n)
      print(input_n, input_numbers)

      if max_pairwise_product(input_numbers) == max_pairwise_product_quick(input_numbers):
        print(max_pairwise_product_quick(input_numbers), "OK")
        counter += 1
      else:
        print(max_pairwise_product(input_numbers))
        print(max_pairwise_product_quick(input_numbers))
        print("DIFFERENT")
        break


```
